## Import The Neccessary Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read The CSV File Data

```python
df = pd.read_csv('train.csv')
df.head(10)
```

```python
# Dropping the Country as its remains constant for all the record
# Dropping the Row ID because its different for all the records
# Droping the Customer ID and Product ID because the they are generally unique for every customer and product respectively
df.drop(['Row ID', 'Country', 'Customer ID', 'Product ID'], axis =1 , inplace = True)
```

## Visualize The Sales In The States

```python
Per_state_sales = df.groupby("State")['Sales'].sum().reset_index()
plt.figure(figsize=(14,10))
sns.barplot(x = 'Sales', y = 'State', data = Per_state_sales, orient='h', palette='viridis')
```

## Visualize The Sales In All The Regions

```python
sales_per_region = df.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize= (12,8))
sns.barplot(x = 'Region', y = 'Sales', data= sales_per_region, palette='viridis')
```
